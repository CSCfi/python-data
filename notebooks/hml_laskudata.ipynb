# data wrangling with pandas

```python
import pandas as pd
accounts = pd.read_csv("./data/hml/Kirjaussuunnitelma 2014 tili.csv", 
                               encoding="iso-8859-15", sep=";", decimal=",", dtype=str) 
cost_obj = pd.read_csv("./data/hml/Kirjaussuunnitelma 2014 kp.csv", 
                               encoding="iso-8859-15", sep=";", decimal=",", dtype=str) 
parse_lookup = {
                 "Tot eurot": float, 
                 "Docid": str, 
                 "Tositenumero": str, 
                 "Kp": str, 
                 "Tili": str
                }
bills = pd.read_csv(
                    "./data/hml/HML ostolaskut 2014 julkaisu final.csv",
                    encoding="iso-8859-15", 
                    sep=";", 
                    decimal=",", 
                    dtype=parse_lookup
                   ) 
```

```python
accounts.describe()
```

```python

all_ = pd.merge(bills, accounts, how="left", on=["Tili", "Tilin nimi"])

#all_ = all_.drop("Y-tunnus", 1)

all_ = pd.merge(all_, cost_obj, how="left", left_on="Kp", right_on="KP")
all_.describe()
```

```python
#now we can use the
grouped = all_.groupby("Y-tunnus")
sums = grouped.sum().sort_values("Tot eurot", ascending=False)
```
