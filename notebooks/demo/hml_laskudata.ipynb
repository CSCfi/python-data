#data wrangling with pandas

```python
import urllib.request
urllib.request.urlretrieve ("http://www.hameenlinna.fi/pages/563719/"
                           "Ostolaskut/2014/"
                           "HML%20ostolaskut%202014%20julkaisu%20csv.zip",
                           "hml_tilit.zip")
import zipfile
zip_ref = zipfile.ZipFile("hml_tilit.zip", "r")
zip_ref.extractall(".")
zip_ref.close()
                           
```

```python
import pandas as pd
tili = pd.read_csv("Kirjaussuunnitelma 2014 tili.csv", 
                               encoding="iso-8859-15", sep=";", decimal=",") 
kp = pd.read_csv("Kirjaussuunnitelma 2014 kp.csv", 
                               encoding="iso-8859-15", sep=";", decimal=",") 
laskut = pd.read_csv("HML ostolaskut 2014 julkaisu final.csv",
                                encoding="iso-8859-15", sep=";", decimal=",") 
```

```python
laskut.columns
```

```python
all_ = laskut.merge(tili)
all_.Tili = all_.Tili.apply(str)
all_.Kp = all_.Kp.apply(str)
all_.Tositenumero = all_.Tositenumero.apply(str)
all_.Docid = all_.Docid.apply(str)
```

```python
all_["Tili"].dtype
```

```python
grouped = all_.groupby("KP:n nimi")
sums = grouped.sum().sort_values("Tot eurot", ascending=False)
```
